```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset("titanic")
dataset.head()
```

```python
sns.displot(dataset["fare"])
```

```python
sns.displot(dataset["fare"], kde=True)
```

```python
sns.jointplot(x="age", y="fare", data=dataset)
```

```python
sns.displot(dataset["fare"], kde=False, bins=10)
sns.jointplot(x="age", y="fare", data=dataset, kind="hex")
```

```python
#The rugplot() is used to draw small bars along x-axis for each point in the dataset. To plot a rug
#plot, you need to pass the name of the column. Let us plot a rug plot for fare.
sns.rugplot(dataset["fare"])
```
