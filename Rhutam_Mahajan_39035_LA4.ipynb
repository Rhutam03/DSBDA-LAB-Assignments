```python
#Data Analytics I
#Create a Linear Regression Model using Python/R to predict home prices using Boston Housing 
#Dataset (https://www.kaggle.com/c/boston-housing). The Boston Housing dataset contains 
#information about various houses in Boston through different parameters. There are 506 samples 
#and 14 feature variables in this dataset. 
#The objective is to predict the value of prices of the house using the given features.
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
#Load data
df=pd.read_csv("C:\\Users\\rhuta\\Downloads\\boston_housing.csv.csv")
df
```

```python
#View dimension of dataset
df.shape
```

```python
#View some statistics of dataset mean,max,min etc
df.describe()
```

```python
#View dataset features and attribute information
df.head()
```

```python
# Adding target variable to dataframe
df['price'] = boston.target
```

```python
#split dependent and independent variables
x = df.drop(['price'], axis = 1)
y = df['price']
```

```python
#Initialize the linear regression model
reg =linear_model.LinearRegression()
```

```python
#Split the data into 67% training and 33% testing data
#NOTE: We have to split the dependent variables (x) and the target or independent variable (y)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state = 42)
```

```python
#Train our model with the training data
reg.fit(x_train, y_train)
```

```python
#Print the coeffecients/weights for each feature/column of our model
print(reg.coef_)
```

```python
#print our price predictions on our test data
y_pred = reg.predict(x_test)
print(y_pred)
```

```python
#To check the model's performance/accuracy we will use a metric called mean squared error (MSE)
print(np.mean((y_pred-y_test)**2))
```
